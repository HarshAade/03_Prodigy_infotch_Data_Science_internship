```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  # Training Accuracy
    print('Testing Score',test_score)    # Testing Accuracy
```

```python
# Removes patter in the input text

def remove_pattern(input_txt, pattern):
    # Find all occurrences of the pattern in the input text
    r = re.findall(pattern, input_txt)

    # Replace each found pattern with an empty string in the input text
    for word in r:
        input_txt = re.sub(word, "", input_txt)

    return input_txt
```

```python
twitter_df.head()
```

```python
#  Remove twitter handles (@user)
twitter_df['clean_tweet'] = np.vectorize(remove_pattern)(twitter_df['tweet'], "@[\w]*")
```

```python
twitter_df.head()
# In newly Created column clean_tweet you can see we have all the twitter handles "@user"
```

As we have remove " @user " if there are 2 , 3 etc it will remove all the @user

```python
# Removing Special charcters , number & punctuations
twitter_df['clean_tweet'] = twitter_df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
twitter_df.head()
# As you can see, in this example, we have removed some special characters like a semicolon, etc.
# For example, you can check the last row in the below cell; there is a semicolon after "factsguide,"
# and in the newly created column, it is removed.
```

```python
# removing Short words
twitter_df['clean_tweet'] = twitter_df['clean_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w) > 3]))
```

```python
# tokenized = spetraing full sentence into single words  (  Individual words concsidered as token )
tokenized_tweet = twitter_df['clean_tweet'].apply(lambda x: x.split())
tokenized_tweet.head()
```

    Steaming in Machine Learning refers to the process of reducing words to their base or root form.

```python
# Stem the words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda sentence:[stemmer.stem(word)for word in sentence])
tokenized_tweet.head()
```

```python
# Extract the hashtags
def hashtag_extract(tweets):
    hashtags = []
    # Loop through words in the tweet
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)  # Correct the regular expression pattern
        hashtags.extend(ht)
    return hashtags
```

```python
# Feature extraction
from sklearn.feature_extraction.text import CountVectorizer

# Correct the typos in the CountVectorizer parameters
bow_vectorizer = CountVectorizer(max_df=0.90, min_df=1000, stop_words='english')
bow = bow_vectorizer.fit_transform(twitter_df['clean_tweet'])
```
